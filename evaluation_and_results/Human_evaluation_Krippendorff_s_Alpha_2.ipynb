```python
#adjacency matrix
import copy
import numpy as np
import pandas as pd
def kdkd(o):
  return kd.alpha(o, level_of_measurement='interval')
torg = copy.deepcopy(org)
torg_em = copy.deepcopy(org_em) #temp, without affecting the origins
torg_rel = copy.deepcopy(org_rel) #temp
torg_flu = copy.deepcopy(org_flu) #temp
torgs = [org, torg_em, torg_rel, torg_flu] #temp
adorgs = []
for i in range(len(torgs)):
  ads = []
  #calculating each org
  for j in range(len(ttorgs[i])):
    ttorgs = copy.deepcopy(torgs) #temp for temp
    temp = ttorgs[i][j]
    ad = []
    for k in range(len(ttorgs[i])):
      compared_to = ttorgs[i][k]
      ad.append(round(kd.alpha([temp,compared_to]),2))
    ads.append(ad)
  print(ads)
  admt = pd.DataFrame(ads)
  print(admt)
  adorgs.append(ads)
  print(np.shape(adorgs))
```
