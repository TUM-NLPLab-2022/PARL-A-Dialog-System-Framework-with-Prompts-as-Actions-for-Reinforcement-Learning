```python
data = pd.read_csv('generated_responses.csv')
```

# FED Metrics

```python
metrics = ['engaging','relevant','correct','understandable','fluent','coherent','consistent','diverse','likeable','depth','informative','error recovery','flexible','interesting','semantically appropriate','specific']
bots = ['RLPA','blenderbot']
```

```python
for bot in bots:
    for m in metrics:
        data[f'{bot}_{m}'] = 0
```

```python
data
```

```python
from tqdm import tqdm
for i in tqdm(range(len(data))):
    chat_history = data.loc[i,'chat_history']
    chat_history= chat_history.replace('[usr]', '<|endoftext|>')
    chat_history= chat_history.replace('[sys]', '<|endoftext|>')
    chat_history += ' <|endoftext|> '

    for bot in bots:
        conv = chat_history + data.loc[i,f'{bot}_response']
        scores = fed.evaluate(conv,model,tokenizer)
        for m in metrics:
            data.loc[i,f'{bot}_{m}'] = -scores[m]


```

```python
data.to_csv('automatic_evaluation_results.csv')
data
```

```python
n = len(data)
for bot in bots:
    for m in metrics:
        print(f'''{bot}_{m}: {sum(data[f'{bot}_{m}'])/n}''')
```

```python
import evaluate
meteor = evaluate.load("meteor")

def compute_meteor(generated_response, gt_response):
  results = meteor.compute(predictions=[generated_response], references=[gt_response])
  return results['meteor']
```

```python
compute_meteor('Hi how is it going', 'Hi how is it going?')
```

```python
compute_meteor('Oh my gosh, I would have been so scared. What happened? Did they end up running away?', 'Did you suffer any injuries?')
```

```python
from tqdm import tqdm
for i in tqdm(range(len(data))):
  gt = data.loc[i,'gt_response']
  rlpa = data.loc[i,'RLPA_response']
  blenderbot = data.loc[i,'blenderbot_response']

  rlpa_meteor = compute_meteor(rlpa,gt)
  blenderbot_meteor = compute_meteor(blenderbot,gt)

  data.loc[i,'RLPA_meteor'] = rlpa_meteor
  data.loc[i,'blenderbot_meteor'] = blenderbot_meteor
```

```python
n =len(data['RLPA_meteor'])
avg_rpla_meteor = sum(data['RLPA_meteor'])/n
avg_blenderbot = sum(data['blenderbot_meteor'])/n

avg_rpla_meteor,avg_blenderbot
```
